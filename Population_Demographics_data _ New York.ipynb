### Web scrapping of Population data from wikipedia page using BeautifulSoup.
Beautiful Soup is a Python package for parsing HTML and XML documents (including having malformed markup, i.e. non-closed tags, so named after tag soup). It creates a parse tree for parsed pages that can be used to extract data from HTML, which is useful for web scraping.

```python
website_url = requests.get('https://en.wikipedia.org/wiki/Demographics_of_New_York_City').text
soup = BeautifulSoup(website_url,'lxml')
table = soup.find('table',{'class':'wikitable sortable'})
#print(soup.prettify())

headers = [header.text for header in table.find_all('th')]

table_rows = table.find_all('tr')        
rows = []
for row in table_rows:
   td = row.find_all('td')
   row = [row.text for row in td]
   rows.append(row)

with open('BON2_POPULATION1.csv', 'w') as f:
   writer = csv.writer(f)
   writer.writerow(headers)
   writer.writerows(row for row in rows if row)
```

### Loading  data from CSV file

```python
New_data=pd.read_csv('BON2_POPULATION1.csv')
New_data.drop(New_data.columns[[7,8,9,10,11]], axis=1,inplace=True)
print('Data  has been downloaded succesfully!!!')
```

Remove whitespaces and rename columns

```python
New_data.columns = New_data.columns.str.replace(' ', '')
New_data.columns = New_data.columns.str.replace('\'','')
New_data.rename(columns={'Borough':'persons_sq_mi','County':'persons_sq_km'}, inplace=True)
New_data
```

```python
New_data.rename(columns = {'NewYorkCitysfiveboroughsvte\n' : 'Borough',
                   'Jurisdiction\n':'County',
                   'Population\n':'Estimate_2017', 
                   'Landarea\n':'square_miles',
                    'Density\n':'square_km'}, inplace=True)
New_data
```

Replace newline('\n') from each string from left and right sides

```python
New_data['Borough']=New_data['Borough'].replace(to_replace='\n', value='', regex=True)
New_data['County']=New_data['County'].replace(to_replace='\n', value='', regex=True)
New_data['Estimate_2017']=New_data['Estimate_2017'].replace(to_replace='\n', value='', regex=True)
New_data['square_miles']=New_data['square_miles'].replace(to_replace='\n', value='', regex=True)
New_data['square_km']=New_data['square_km'].replace(to_replace='\n', value='', regex=True)
New_data['persons_sq_mi']=New_data['persons_sq_mi'].replace(to_replace='\n', value='', regex=True)
#New_data['GrossDomesticProduct']=New_data['GrossDomesticProduct'].replace(to_replace='\n', value='', regex=True)
New_data
```

```python
New_data.drop(["squarekm", "persons/sq.mi"] , axis = 1, inplace = True)
```

Shift data in the last two rows

```python
New_data.loc[5:,['persons_sq_mi']] = New_data.loc[2:,['persons_sq_mi']].shift(1,axis=1)
New_data.loc[5:,['square_km','persons_sq_mi']] = New_data.loc[2:,['square_km','persons_sq_mi']].shift(1,axis=1)
New_data.loc[5:,['square_miles','square_km']] = New_data.loc[2:,['square_miles','square_km']].shift(1,axis=1)
New_data.loc[5:,['Estimate_2017','square_miles']] = New_data.loc[2:,['Estimate_2017','square_miles']].shift(1,axis=1)
New_data.loc[5:,['County','Estimate_2017']] = New_data.loc[2:,['County','Estimate_2017']].shift(1,axis=1)
New_data.loc[5:,['Borough','County']] = New_data.loc[2:,['Borough','County']].shift(1,axis=1)
New_data.head()
```
